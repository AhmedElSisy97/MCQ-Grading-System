```python
import cv2 as cv
```

```python
from matplotlib import image as mpi
```

```python
from matplotlib import pyplot as plt
```

```python
coins_colored = cv.imread('coins2.jpg')
```

```python
#converting to grayscale
coins = cv.cvtColor(coins_colored, cv.COLOR_BGR2GRAY)
```

```python
#show input image
plt.imshow(coins, cmap='gray')
plt.show()
```

```python
#apply midean blur to reduce light reflection
```

```python
median = cv.medianBlur(coins, 5)
```

```python
#histogram to get the gray range of coins
hist = cv.calcHist([median], [0], None, [256], [0, 256])
plt.hist(median.ravel(), 256, [0, 256])
plt.title('Histogram of Coins')
plt.show()
```

```python
#thresholding
_, thresh = cv.threshold(median, 90, 255, cv.THRESH_BINARY)
```

```python
#thresholded image
```

```python
plt.imshow(thresh, cmap = 'gray')
plt.title('thresholded image')
plt.show()
```

```python
#finding contours in image
contours, _ = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
```

```python
#creating a list of areas of the contours
areas = []
for cont in contours:
    area = cv.contourArea(cont)
    areas.append(area)
areas.sort()
print(areas)
```

```python
'''
from list we find that the max area of 50 cent coins
in this image is 7021.5 and any coin bigger than that
is a 100 cent
'''
```

```python
#drawing contours on the original image red for 100 cent and blue for 50 cent
c50 = 0
c100 = 0
for cont in contours:
    area = cv.contourArea(cont)
    #drawing red contours and counting 100c 
    if area > 7022:
        cv.drawContours(coins_colored, cont, -1, (0, 0, 255), 4)
        c100 += 1
    #drawing blue contours and counting 50c
    else:
        cv.drawContours(coins_colored, cont, -1, (255, 0, 0), 4)
        c50 += 1
```

```python
#convert BGR to RGB
RGB = cv.cvtColor(coins_colored, cv.COLOR_BGR2RGB)
```

```python
#show result
plt.imshow(RGB)
plt.title('OUTPUT')
plt.show()
```

```python
#coins count
print('Number of 100c coins= ' + c100.__str__() + '\n')
print('Number of 50c coins= ' + c50.__str__() + '\n')
print('Total Value= '+ (100*c100 + 50*c50).__str__() + ' cents\n')
```
