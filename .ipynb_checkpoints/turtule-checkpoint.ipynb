```python
MO=(10,50)
t = turtle.Turtle()

t.goto(MO)

turtle.done()
```

```python
def box(po1,po3): #p is the point ,example p1=(20,50)
    X1=po1[0]
    Y1=po1[1]
    X3=po3[0]
    Y3=po3[1]
    po2=(X3,Y1) 
    po4=(X1,Y3)
    t = turtle.Turtle()
    t.goto(po1)
    t.goto(po2)
    t.goto(po3)
    t.goto(po4)
    t.goto(po1)
    turtle.done()
    return t

```

```python
R=(100,50)
T=(50,100)

box(R,T)
```
